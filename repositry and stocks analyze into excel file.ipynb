```python
import pandas as pd
from datetime import datetime

# Load initial inventory (student data)
inventory = pd.read_csv("data.csv")

def view_inventory():
    """View the current inventory of student data."""
    print("\nCurrent Student Information:")
    print(inventory)

def add_to_inventory():
    """Add or update student records."""
    global inventory  # Declaring `inventory` as global to modify it

    rollno = input("\nEnter Roll Number: ").strip()
    result = input("Enter result: ").strip()
    year = input("Enter year: ").strip()
    
    # Update if student exists, otherwise add a new record
    if rollno in inventory['Rollno'].values:
        inventory.loc[inventory['Rollno'] == rollno, 'year'] = year
        inventory.loc[inventory['Rollno'] == rollno, 'result'] = result
        print(f"Updated information for Roll Number {rollno}.")
    else:
        new_row = pd.DataFrame({'Rollno': [rollno], 'result': [result], 'year': [year]})
        inventory = pd.concat([inventory, new_row], ignore_index=True)
        print(f"Added new record for Roll Number {rollno}.")
    
    # Save updated inventory back to the CSV file
    inventory.to_csv('data.csv')


def menu():
    """Display the main menu for the system."""
    while True:
        print("\nStudent Management System:")
        print("1. View Student Information")
        print("2. Add or Update Student Records")
        print("3. Exit")
        
        choice = input("\nEnter your choice: ").strip()
        
        if choice == '1':
            view_inventory()
        elif choice == '2':
            add_to_inventory()
        elif choice == '3':
            print("Exiting the Student Management System. Goodbye!")
            break
        else:
            print("Invalid choice, please try again.")

# Run the menu
menu()
```

```python
def sqrt_iterative(number, max_iter=100,tolerance=1e-6):
    if number<0:
       return ValueError("apka ne jo number enter kiya hai bo negative number hai to gadhe  eska squareroot nhi hota byy lallu")
    x = number 
    for _ in range(max_iter):
     next_x = 0.5 * (x + number / x) 
     if abs(next_x - x)<tolerance:
       return next_x
    x = next_x

    return  x
    
result = sqrt_iterative(5)
print("Approximate Square Root:", result)
       
```

```python
def analyze_stock_prices(prices):

 max_price =max(prices)
 
 min_price =min(prices)
 
 avg_price =sum(prices)/len(prices)
 
 return max_price, min_price, avg_price
 
tcs_stock_prices = pd.read_excel("file.xlsx")

prices_column = tcs_stock_prices['tcs stock']

prices_column = pd.to_numeric(prices_column, errors='coerce').dropna()
prices=prices_column.tolist()

print("tcs stock prices of a week",prices)

max_price, min_price, avg_price= analyze_stock_prices(prices)

print(f"Maximum Price: {max_price} INR")
print(f"Minimum Price: {min_price} INR")
print(f"Average Price: {avg_price:.2f} INR")

 
```
