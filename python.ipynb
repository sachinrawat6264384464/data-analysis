```python
import pandas as pd
from datetime import datetime

# Initial inventory


# Sales log
sales_log = pd.DataFrame(columns=['Date', 'Customer', 'Item', 'Quantity', 'Total'])

# Customer database
customers = pd.DataFrame(columns=['Name', 'Phone', 'Email', 'Total_Spent'])

def view_inventory():
    print("\nCurrent Inventory:")
    print(inventory)

def add_to_inventory():
    item = input("\nEnter item name: ").strip()
    price = float(input("Enter price of the item: "))
    stock = int(input("Enter quantity: "))
    
    # Update if item exists, otherwise add a new row
    if item in inventory['Item'].values:
        inventory.loc[inventory['Item'] == item, 'Stock'] += stock
        inventory.loc[inventory['Item'] == item, 'Price'] = price
    else:
         new_row = {'Item': item, 'Price': price, 'Stock': stock}
         inventory.loc[len(inventory)] = new_row
    print(f"{item} has been added/updated in the inventory.")

def sell_item():
    view_inventory()
    customer_name = input("\nEnter customer name: ").strip()
    customer_phone = input("Enter customer phone: ").strip()
    customer_email = input("Enter customer email: ").strip()

    # Add customer to database if new
    if customer_name not in customers['Name'].values:
        customers.loc[len(customers)] = [customer_name, customer_phone, customer_email, 0]
    
    item = input("\nEnter item to sell: ").strip()
    if item not in inventory['Item'].values:
        print(f"Error: {item} not found in inventory.")
        return

    quantity = int(input("Enter quantity to sell: "))
    stock_available = inventory.loc[inventory['Item'] == item, 'Stock'].values[0]
    
    if quantity > stock_available:
        print(f"Error: Only {stock_available} items available.")
        return

    total_price = inventory.loc[inventory['Item'] == item, 'Price'].values[0] * quantity
    
    # Update stock
    inventory.loc[inventory['Item'] == item, 'Stock'] -= quantity

    # Update customer's total spent
    customers.loc[customers['Name'] == customer_name, 'Total_Spent'] += total_price
    
    # Log sale
    global sales_log
    sales_log.loc[len(sales_log)] = [
        datetime.now().strftime('%Y-%m-%d %H:%M:%S'),
        customer_name,
        item,
        quantity,
        total_price
    ]
    
    print(f"Sold {quantity} {item}(s) to {customer_name} for ${total_price:.2f}.")

def view_sales():
    print("\nSales Log:")
    print(sales_log)

def view_customers():
    print("\nCustomer Database:")
    print(customers)

def calculate_total_revenue():
    total_revenue = sales_log['Total'].sum()
    print(f"\nTotal Revenue: ${total_revenue:.2f}")

def menu():
    while True:
        print("\nBakery Management System:")
        print("1. View Inventory")
        print("2. Add to Inventory")
        print("3. Sell Item")
        print("4. View Sales")
        print("5. View Customers")
        print("6. Calculate Total Revenue")
        print("7. Exit")
        
        choice = input("\nEnter your choice: ").strip()
        
        if choice == '1':
            view_inventory()
        elif choice == '2':
            add_to_inventory()
        elif choice == '3':
            sell_item()
        elif choice == '4':
            view_sales()
        elif choice == '5':
            view_customers()
        elif choice == '6':
            calculate_total_revenue()
        elif choice == '7':
            print("Exiting the Bakery Management System. Goodbye!")
            break
        else:
            print("Invalid choice, please try again.")

# Run the menu
menu()
```
